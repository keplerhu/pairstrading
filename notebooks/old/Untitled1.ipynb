```python
import yfinance as yf
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# === FIXED PARAMETERS ===
beta = 1.3856  # from Engle-Granger
window = 20
ticker1, ticker2 = "UMAC", "RCAT"

# === LOAD DATA ===
data = yf.download([ticker1, ticker2], period="1y", interval="1d")['Close'].dropna()
spread = data[ticker1] - beta * data[ticker2]

# === DEFINE OPTIMIZATION GRID ===
entry_vals = np.arange(1.0, 2.1, 0.1)
exit_vals = np.arange(0.1, 0.6, 0.05)

results = []

# === GRID SEARCH ===
for z_entry in entry_vals:
    for z_exit in exit_vals:
        if z_exit >= z_entry:
            continue  # skip invalid combinations

        df = pd.DataFrame(index=data.index)
        df['spread'] = spread
        df['mean'] = df['spread'].rolling(window).mean()
        df['std'] = df['spread'].rolling(window).std()
        df['zscore'] = (df['spread'] - df['mean']) / df['std']

        df['position'] = 0
        df.loc[df['zscore'] > z_entry, 'position'] = -1
        df.loc[df['zscore'] < -z_entry, 'position'] = 1
        df.loc[df['zscore'].abs() < z_exit, 'position'] = 0
        df['position'] = df['position'].replace(0, np.nan).ffill()

                # === Add slippage + trade logging ===
        slippage_rate = 0.001  # 0.1% per side

                # === Detect trade entries/exits ===
        df['signal'] = df['position'].diff()
        
        trade_log = []
        in_trade = False
        
        for i in range(1, len(df)):
            row = df.iloc[i]
            idx = df.index[i]
        
            # ENTRY
            if row['signal'] != 0 and not in_trade:
                in_trade = True
                entry_date = idx
                entry_price = row['spread']
                entry_z = row['zscore']
                entry_pos = row['position']
        
            # EXIT
            if in_trade and row['position'] == 0:
                exit_date = idx
                exit_price = row['spread']
                exit_z = row['zscore']
                direction = 'Long Spread' if entry_pos == 1 else 'Short Spread'
                raw_pnl = (exit_price - entry_price) * entry_pos
                cost = slippage_rate * abs(entry_price) + slippage_rate * abs(exit_price)
                net_pnl = raw_pnl - cost
                holding_period = (exit_date - entry_date).days
        
                trade_log.append({
                    'Entry Date': entry_date,
                    'Exit Date': exit_date,
                    'Direction': direction,
                    'Entry Z': entry_z,
                    'Exit Z': exit_z,
                    'Gross PnL': raw_pnl,
                    'Net PnL': net_pnl,
                    'Holding Period': holding_period
                })
        
                in_trade = False




        df['spread_return'] = df['spread'].diff()
        df['strategy_return'] = df['position'].shift(1) * df['spread_return']

        daily_returns = df['strategy_return'].dropna()
        cum_return = df['strategy_return'].cumsum().iloc[-1]
        annual_return = daily_returns.mean() * 252
        annual_vol = daily_returns.std() * np.sqrt(252)
        sharpe = annual_return / annual_vol if annual_vol > 0 else np.nan

        entries = df['position'].diff().fillna(0).abs()
        num_trades = int(entries.sum() / 2)

        results.append({
            'z_entry': z_entry,
            'z_exit': z_exit,
            'sharpe': sharpe,
            'cumulative_return': cum_return,
            'trades': num_trades
        })
        if sharpe > 2.9:
            trade_df = pd.DataFrame(trade_log)
            fname = f"trade_log_z{z_entry:.2f}_x{z_exit:.2f}.csv"
            trade_df.to_csv(fname, index=False)
            print(f"Saved trade log to: {fname}")



# === DISPLAY RESULTS ===
results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by='sharpe', ascending=False)

print("=== Top Strategy Configurations (by Sharpe Ratio) ===")
print(results_df.head(10).to_string(index=False))
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

# === PARAMETERS ===
ticker1, ticker2 = "UMAC", "RCAT"
rolling_window_beta = 60  # for rolling OLS hedge ratio
zscore_window = 20
z_entry = 1.4
z_exit = 0.2
slippage_rate = 0.001  # 0.1% per leg

# === LOAD DATA ===
data = yf.download([ticker1, ticker2], period="1y", interval="1d")['Close'].dropna()

# === COMPUTE ROLLING HEDGE RATIO (BETA) ===
betas = [np.nan] * rolling_window_beta

for i in range(rolling_window_beta, len(data)):
    y = data[ticker1].iloc[i - rolling_window_beta:i]
    x = data[ticker2].iloc[i - rolling_window_beta:i]
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    betas.append(model.params[1])

data['beta'] = betas

# === SPREAD AND Z-SCORE ===
data['spread'] = data[ticker1] - data['beta'] * data[ticker2]
data['mean'] = data['spread'].rolling(zscore_window).mean()
data['std'] = data['spread'].rolling(zscore_window).std()
data['zscore'] = (data['spread'] - data['mean']) / data['std']

# === SIGNALS ===
data['position'] = 0
data.loc[data['zscore'] > z_entry, 'position'] = -1
data.loc[data['zscore'] < -z_entry, 'position'] = 1
data.loc[data['zscore'].abs() < z_exit, 'position'] = np.nan
data['position'] = data['position'].ffill()

# === STRATEGY RETURNS ===
data['spread_return'] = data['spread'].diff()
data['strategy_return'] = data['position'].shift(1) * data['spread_return']

# === APPLY SLIPPAGE (on entry + exit points) ===
data['signal'] = data['position'].diff()
trade_costs = (data['signal'].abs() * data['spread'].abs() * slippage_rate).fillna(0)
data['strategy_return_net'] = data['strategy_return'] - trade_costs
data['cum_strategy_return'] = data['strategy_return_net'].cumsum()

# === METRICS ===
daily_returns = data['strategy_return_net'].dropna()
cum_return = data['cum_strategy_return'].iloc[-1]
annual_return = daily_returns.mean() * 252
annual_vol = daily_returns.std() * np.sqrt(252)
sharpe = annual_return / annual_vol if annual_vol > 0 else np.nan

rolling_max = data['cum_strategy_return'].cummax()
drawdown = data['cum_strategy_return'] - rolling_max
max_dd = drawdown.min()

entries = data['position'].diff().fillna(0).abs()
num_trades = int(entries.sum() / 2)

# === OUTPUT ===
print("=== Strategy Performance Summary with Rolling Beta ===")
print(f"Z-entry threshold:     ±{z_entry}")
print(f"Z-exit threshold:      < {z_exit}")
print(f"Cumulative Return:     {cum_return:.4f}")
print(f"Annualized Return:     {annual_return:.4f}")
print(f"Annualized Volatility: {annual_vol:.4f}")
print(f"Sharpe Ratio:          {sharpe:.2f}")
print(f"Max Drawdown:          {max_dd:.4f}")
print(f"Number of Trades:      {num_trades}")

# === PLOTS ===
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(data['spread'], label='Spread')
plt.plot(data['mean'], '--', label='Rolling Mean')
plt.title('Rolling Spread and Mean')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(data['cum_strategy_return'], label='Cumulative Net Return')
plt.title('Strategy Cumulative Return (Net of Slippage)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(data['beta'], label='Rolling Beta')
plt.axhline(y=1.3856, color='r', linestyle='--', label='Fixed Beta')
plt.title('Rolling vs Fixed Beta')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import yfinance as yf
from itertools import product

# === CONFIGURATION ===
ticker1, ticker2 = "UMAC", "RCAT"
beta_fixed = 1.3856
capital = 10000
slippage_rate = 0.001
train_end = '2025-03-31'
test_start = '2025-04-01'

# === DOWNLOAD DATA ===
data = yf.download([ticker1, ticker2], period="1y", interval="1d")['Close'].dropna()

# === SPLIT TRAIN/TEST ===
train = data.loc[:train_end].copy()
test = data.loc[test_start:].copy()

# === STRATEGY FUNCTION ===
def backtest(df, z_entry, z_exit, z_window, beta, capital, slippage):
    spread = df[ticker1] - beta * df[ticker2]
    mean = spread.rolling(z_window).mean()
    std = spread.rolling(z_window).std()
    zscore = (spread - mean) / std

    position = pd.Series(index=df.index, dtype=int)
    position[zscore > z_entry] = -1
    position[zscore < -z_entry] = 1
    position[zscore.abs() < z_exit] = np.nan
    position = position.ffill().fillna(0).astype(int)

    units = capital / (df[ticker1].abs() + beta * df[ticker2].abs())
    spread_return = spread.diff()
    strategy_return = position.shift(1) * spread_return * units.shift(1)

    signal = position.diff().fillna(0)
    slip_cost = signal.abs() * (df[ticker1].abs() + beta * df[ticker2].abs()) * slippage
    net_return = strategy_return - slip_cost

    return_pct = net_return / capital
    cum_return_pct = return_pct.cumsum()

    daily_ret = return_pct.dropna()
    ann_return = daily_ret.mean() * 252
    ann_vol = daily_ret.std() * np.sqrt(252)
    sharpe = ann_return / ann_vol if ann_vol > 0 else np.nan
    max_dd = (cum_return_pct - cum_return_pct.cummax()).min()

    entries = position.diff().fillna(0).abs()
    num_trades = int(entries.sum() / 2)

    metrics = {
        'z_entry': z_entry, 'z_exit': z_exit, 'z_window': z_window,
        'sharpe': sharpe, 'cumulative_return': cum_return_pct.iloc[-1],
        'max_drawdown': max_dd, 'trades': num_trades,
        'returns': cum_return_pct
    }
    return metrics

# === PARAMETER GRID ===
z_entry_range = np.arange(1.0, 2.1, 0.1)
z_exit_range = np.arange(0.0, 1.1, 0.1)
z_window_range = np.arange(10, 31, 5)

results = []

# === OPTIMIZE ON TRAINING SET ===
for z_entry, z_exit, z_window in product(z_entry_range, z_exit_range, z_window_range):
    if z_exit >= z_entry:
        continue
    metrics = backtest(train, z_entry, z_exit, z_window, beta_fixed, capital, slippage_rate)
    results.append(metrics)

results_df = pd.DataFrame(results).sort_values(by="sharpe", ascending=False).reset_index(drop=True)
best_params = results_df.iloc[0]

# === APPLY BEST STRATEGY TO TEST SET ===
test_metrics = backtest(test, best_params.z_entry, best_params.z_exit, best_params.z_window, beta_fixed, capital, slippage_rate)

# === DISPLAY RESULTS ===
print("=== Optimal Parameters on Training Set ===")
print(best_params[['z_entry', 'z_exit', 'z_window']])
print("\n=== Performance on Training Set ===")
print(f"Sharpe Ratio:          {best_params.sharpe:.2f}")
print(f"Cumulative Return:     {best_params.cumulative_return * 100:.2f}%")
print(f"Max Drawdown:          {best_params.max_drawdown * 100:.2f}%")
print(f"Number of Trades:      {best_params.trades}")

print("\n=== Performance on Test Set ===")
print(f"Sharpe Ratio:          {test_metrics['sharpe']:.2f}")
print(f"Cumulative Return:     {test_metrics['cumulative_return'] * 100:.2f}%")
print(f"Max Drawdown:          {test_metrics['max_drawdown'] * 100:.2f}%")
print(f"Number of Trades:      {test_metrics['trades']}")

# === PLOT RETURNS ===
plt.figure(figsize=(14, 6))
plt.plot(train.index, best_params.returns * 100, label="Train")
plt.plot(test.index, test_metrics['returns'] * 100, label="Test")
plt.title("Cumulative Strategy Return (%)")
plt.xlabel("Date")
plt.ylabel("Cumulative Return (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from itertools import product

# === CONFIG ===
ticker1, ticker2 = "UMAC", "RCAT"
beta_fixed = 1.3856
capital = 10000
slippage_rate = 0.001
train_end = '2025-03-31'
test_start = '2025-04-01'

# === DOWNLOAD DATA ===
data = yf.download([ticker1, ticker2], period="1y", interval="1d")['Close'].dropna()
train = data.loc[:train_end].copy()
test = data.loc[test_start:].copy()

# === BACKTEST FUNCTION (NON-COMPOUNDING, TRACK POSITIONS) ===
def backtest(df, z_entry, z_exit, z_window, beta, capital, slippage):
    spread = df[ticker1] - beta * df[ticker2]
    mean = spread.rolling(z_window).mean()
    std = spread.rolling(z_window).std()
    zscore = (spread - mean) / std

    # Generate signals
    position = pd.Series(index=df.index, dtype=int)
    position[zscore > z_entry] = -1
    position[zscore < -z_entry] = 1
    position[zscore.abs() < z_exit] = np.nan
    position = position.ffill().fillna(0).astype(int)

    # Units = fixed capital every trade
    trade_value = df[ticker1].abs() + beta * df[ticker2].abs()
    units = capital / trade_value
    spread_return = spread.diff()
    raw_pnl = position.shift(1) * spread_return * units.shift(1)

    # Slippage costs on signal change
    signal = position.diff().fillna(0)
    slip_cost = signal.abs() * trade_value * slippage
    net_pnl = raw_pnl - slip_cost

    # Position-level tracking
    trade_log = pd.DataFrame(index=df.index)
    trade_log['position'] = position
    trade_log['spread'] = spread
    trade_log['pnl'] = net_pnl
    trade_log['cum_pnl'] = net_pnl.cumsum()
    trade_log['return_pct'] = net_pnl / capital
    trade_log['cum_return_pct'] = trade_log['return_pct'].cumsum()
    trade_log['signal'] = signal

    # Metrics
    daily_ret = trade_log['return_pct'].dropna()
    ann_return = daily_ret.mean() * 252
    ann_vol = daily_ret.std() * np.sqrt(252)
    sharpe = ann_return / ann_vol if ann_vol > 0 else np.nan
    max_dd = (trade_log['cum_return_pct'] - trade_log['cum_return_pct'].cummax()).min()
    num_trades = ((signal != 0) & (position.shift(1) == 0)).sum()

    return {
        'z_entry': z_entry, 'z_exit': z_exit, 'z_window': z_window,
        'sharpe': sharpe,
        'cumulative_return': trade_log['cum_return_pct'].iloc[-1],
        'max_drawdown': max_dd,
        'trades': num_trades,
        'returns': trade_log['cum_return_pct'],
        'log': trade_log
    }

# === PARAMETER GRID ===
z_entry_range = np.arange(1.0, 2.1, 0.1)
z_exit_range = np.arange(0.0, 1.1, 0.1)
z_window_range = np.arange(10, 31, 5)

results = []
for z_entry, z_exit, z_window in product(z_entry_range, z_exit_range, z_window_range):
    if z_exit >= z_entry:
        continue
    metrics = backtest(train, z_entry, z_exit, z_window, beta_fixed, capital, slippage_rate)
    results.append(metrics)

# Sort by Sharpe and select best
results_df = pd.DataFrame(results).sort_values(by="sharpe", ascending=False).reset_index(drop=True)
best_params = results_df.iloc[0]
best_train_log = results_df.loc[0, 'log']

# === APPLY TO TEST SET ===
test_result = backtest(test, best_params.z_entry, best_params.z_exit, best_params.z_window, beta_fixed, capital, slippage_rate)
test_log = test_result['log']

# === DISPLAY ===
print("=== Optimal Parameters on Training Set (Fixed Capital, Position Tracked) ===")
print(best_params[['z_entry', 'z_exit', 'z_window']])
print("\n=== Training Set Performance ===")
print(f"Sharpe Ratio:          {best_params.sharpe:.2f}")
print(f"Cumulative Return:     {best_params.cumulative_return * 100:.2f}%")
print(f"Max Drawdown:          {best_params.max_drawdown * 100:.2f}%")
print(f"Number of Trades:      {best_params.trades}")

print("\n=== Test Set Performance ===")
print(f"Sharpe Ratio:          {test_result['sharpe']:.2f}")
print(f"Cumulative Return:     {test_result['cumulative_return'] * 100:.2f}%")
print(f"Max Drawdown:          {test_result['max_drawdown'] * 100:.2f}%")
print(f"Number of Trades:      {test_result['trades']}")

# === PLOT ===
plt.figure(figsize=(14, 6))
plt.plot(best_train_log['cum_return_pct'] * 100, label="Train")
plt.plot(test_log['cum_return_pct'] * 100, label="Test")
plt.title("Cumulative Return (%), Fixed Capital (Non-Compounding)")
plt.xlabel("Date")
plt.ylabel("Cumulative Return (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
