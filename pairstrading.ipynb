```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import yfinance as yf
from itertools import product

# === CONFIGURATION ===
ticker1, ticker2 = "UMAC", "RCAT"
beta_fixed = 1.3856
capital = 10000
slippage_rate = 0.001
train_end = '2025-03-31'
test_start = '2025-04-01'

# === DOWNLOAD DATA ===
data = yf.download([ticker1, ticker2], period="1y", interval="1d")['Close'].dropna()

# === SPLIT TRAIN/TEST ===
train = data.loc[:train_end].copy()
test = data.loc[test_start:].copy()

# === STRATEGY FUNCTION ===
def backtest(df, z_entry, z_exit, z_window, beta, capital, slippage):
    spread = df[ticker1] - beta * df[ticker2]
    mean = spread.rolling(z_window).mean()
    std = spread.rolling(z_window).std()
    zscore = (spread - mean) / std

    position = pd.Series(index=df.index, dtype=int)
    position[zscore > z_entry] = -1
    position[zscore < -z_entry] = 1
    position[zscore.abs() < z_exit] = np.nan
    position = position.ffill().fillna(0).astype(int)

    units = capital / (df[ticker1].abs() + beta * df[ticker2].abs())
    spread_return = spread.diff()
    strategy_return = position.shift(1) * spread_return * units.shift(1)

    signal = position.diff().fillna(0)
    slip_cost = signal.abs() * (df[ticker1].abs() + beta * df[ticker2].abs()) * slippage
    net_return = strategy_return - slip_cost

    return_pct = net_return / capital
    cum_return_pct = return_pct.cumsum()

    daily_ret = return_pct.dropna()
    ann_return = daily_ret.mean() * 252
    ann_vol = daily_ret.std() * np.sqrt(252)
    sharpe = ann_return / ann_vol if ann_vol > 0 else np.nan
    max_dd = (cum_return_pct - cum_return_pct.cummax()).min()

    entries = position.diff().fillna(0).abs()
    num_trades = int(entries.sum() / 2)

    metrics = {
        'z_entry': z_entry, 'z_exit': z_exit, 'z_window': z_window,
        'sharpe': sharpe, 'cumulative_return': cum_return_pct.iloc[-1],
        'max_drawdown': max_dd, 'trades': num_trades,
        'returns': cum_return_pct
    }
    return metrics

# === PARAMETER GRID ===
z_entry_range = np.arange(1.0, 2.1, 0.1)
z_exit_range = np.arange(0.0, 1.1, 0.1)
z_window_range = np.arange(10, 31, 5)

results = []

# === OPTIMIZE ON TRAINING SET ===
for z_entry, z_exit, z_window in product(z_entry_range, z_exit_range, z_window_range):
    if z_exit >= z_entry:
        continue
    metrics = backtest(train, z_entry, z_exit, z_window, beta_fixed, capital, slippage_rate)
    results.append(metrics)

results_df = pd.DataFrame(results).sort_values(by="sharpe", ascending=False).reset_index(drop=True)
best_params = results_df.iloc[0]

# === APPLY BEST STRATEGY TO TEST SET ===
test_metrics = backtest(test, best_params.z_entry, best_params.z_exit, best_params.z_window, beta_fixed, capital, slippage_rate)

# === DISPLAY RESULTS ===
print("=== Optimal Parameters on Training Set ===")
print(best_params[['z_entry', 'z_exit', 'z_window']])
print("\n=== Performance on Training Set ===")
print(f"Sharpe Ratio:          {best_params.sharpe:.2f}")
print(f"Cumulative Return:     {best_params.cumulative_return * 100:.2f}%")
print(f"Max Drawdown:          {best_params.max_drawdown * 100:.2f}%")
print(f"Number of Trades:      {best_params.trades}")

print("\n=== Performance on Test Set ===")
print(f"Sharpe Ratio:          {test_metrics['sharpe']:.2f}")
print(f"Cumulative Return:     {test_metrics['cumulative_return'] * 100:.2f}%")
print(f"Max Drawdown:          {test_metrics['max_drawdown'] * 100:.2f}%")
print(f"Number of Trades:      {test_metrics['trades']}")

# === PLOT RETURNS ===
plt.figure(figsize=(14, 6))
plt.plot(train.index, best_params.returns * 100, label="Train")
plt.plot(test.index, test_metrics['returns'] * 100, label="Test")
plt.title("Cumulative Strategy Return (%)")
plt.xlabel("Date")
plt.ylabel("Cumulative Return (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from itertools import product

# === CONFIG ===
ticker1, ticker2 = "UMAC", "RCAT"
beta_fixed = 1.3856
capital = 10000
slippage_rate = 0.001
train_end = '2025-03-31'
test_start = '2025-04-01'

# === DOWNLOAD DATA ===
data = yf.download([ticker1, ticker2], period="1y", interval="1d")['Close'].dropna()
train = data.loc[:train_end].copy()
test = data.loc[test_start:].copy()

# === BACKTEST FUNCTION (NON-COMPOUNDING, TRACK POSITIONS) ===
def backtest(df, z_entry, z_exit, z_window, beta, capital, slippage):
    spread = df[ticker1] - beta * df[ticker2]
    mean = spread.rolling(z_window).mean()
    std = spread.rolling(z_window).std()
    zscore = (spread - mean) / std

    # Generate signals
    position = pd.Series(index=df.index, dtype=int)
    position[zscore > z_entry] = -1
    position[zscore < -z_entry] = 1
    position[zscore.abs() < z_exit] = np.nan
    position = position.ffill().fillna(0).astype(int)

    # Units = fixed capital every trade
    trade_value = df[ticker1].abs() + beta * df[ticker2].abs()
    units = capital / trade_value
    spread_return = spread.diff()
    raw_pnl = position.shift(1) * spread_return * units.shift(1)

    # Slippage costs on signal change
    signal = position.diff().fillna(0)
    slip_cost = signal.abs() * trade_value * slippage
    net_pnl = raw_pnl - slip_cost

    # Position-level tracking
    trade_log = pd.DataFrame(index=df.index)
    trade_log['position'] = position
    trade_log['spread'] = spread
    trade_log['pnl'] = net_pnl
    trade_log['cum_pnl'] = net_pnl.cumsum()
    trade_log['return_pct'] = net_pnl / capital
    trade_log['cum_return_pct'] = trade_log['return_pct'].cumsum()
    trade_log['signal'] = signal

    # Metrics
    daily_ret = trade_log['return_pct'].dropna()
    ann_return = daily_ret.mean() * 252
    ann_vol = daily_ret.std() * np.sqrt(252)
    sharpe = ann_return / ann_vol if ann_vol > 0 else np.nan
    max_dd = (trade_log['cum_return_pct'] - trade_log['cum_return_pct'].cummax()).min()
    num_trades = ((signal != 0) & (position.shift(1) == 0)).sum()

    return {
        'z_entry': z_entry, 'z_exit': z_exit, 'z_window': z_window,
        'sharpe': sharpe,
        'cumulative_return': trade_log['cum_return_pct'].iloc[-1],
        'max_drawdown': max_dd,
        'trades': num_trades,
        'returns': trade_log['cum_return_pct'],
        'log': trade_log
    }

# === PARAMETER GRID ===
z_entry_range = np.arange(1.0, 2.1, 0.1)
z_exit_range = np.arange(0.0, 1.1, 0.1)
z_window_range = np.arange(10, 31, 5)

results = []
for z_entry, z_exit, z_window in product(z_entry_range, z_exit_range, z_window_range):
    if z_exit >= z_entry:
        continue
    metrics = backtest(train, z_entry, z_exit, z_window, beta_fixed, capital, slippage_rate)
    results.append(metrics)

# Sort by Sharpe and select best
results_df = pd.DataFrame(results).sort_values(by="sharpe", ascending=False).reset_index(drop=True)
best_params = results_df.iloc[0]
best_train_log = results_df.loc[0, 'log']

# === APPLY TO TEST SET ===
test_result = backtest(test, best_params.z_entry, best_params.z_exit, best_params.z_window, beta_fixed, capital, slippage_rate)
test_log = test_result['log']

# === DISPLAY ===
print("=== Optimal Parameters on Training Set (Fixed Capital, Position Tracked) ===")
print(best_params[['z_entry', 'z_exit', 'z_window']])
print("\n=== Training Set Performance ===")
print(f"Sharpe Ratio:          {best_params.sharpe:.2f}")
print(f"Cumulative Return:     {best_params.cumulative_return * 100:.2f}%")
print(f"Max Drawdown:          {best_params.max_drawdown * 100:.2f}%")
print(f"Number of Trades:      {best_params.trades}")

print("\n=== Test Set Performance ===")
print(f"Sharpe Ratio:          {test_result['sharpe']:.2f}")
print(f"Cumulative Return:     {test_result['cumulative_return'] * 100:.2f}%")
print(f"Max Drawdown:          {test_result['max_drawdown'] * 100:.2f}%")
print(f"Number of Trades:      {test_result['trades']}")

# === PLOT ===
plt.figure(figsize=(14, 6))
plt.plot(best_train_log['cum_return_pct'] * 100, label="Train")
plt.plot(test_log['cum_return_pct'] * 100, label="Test")
plt.title("Cumulative Return (%), Fixed Capital (Non-Compounding)")
plt.xlabel("Date")
plt.ylabel("Cumulative Return (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
import yfinance as yf

# === CONFIGURATION ===
ticker1, ticker2 = "UMAC", "RCAT"
beta_fixed = 1.3856
capital = 10000
slippage_rate = 0.001
period = "1y"

# === DOWNLOAD DATA ===
data = yf.download([ticker1, ticker2], period=period, interval="1d")['Close'].dropna()

# === BACKTEST FUNCTION ===
def backtest(df, z_entry, z_exit, z_window, beta, capital, slippage):
    spread = df[ticker1] - beta * df[ticker2]
    mean = spread.rolling(z_window).mean()
    std = spread.rolling(z_window).std()
    zscore = (spread - mean) / std

    position = pd.Series(index=df.index, dtype=int)
    position[zscore > z_entry] = -1
    position[zscore < -z_entry] = 1
    position[zscore.abs() < z_exit] = np.nan
    position = position.ffill().fillna(0).astype(int)

    trade_value = df[ticker1].abs() + beta * df[ticker2].abs()
    units = capital / trade_value
    spread_return = spread.diff()
    raw_pnl = position.shift(1) * spread_return * units.shift(1)

    signal = position.diff().fillna(0)
    slip_cost = signal.abs() * trade_value * slippage
    net_pnl = raw_pnl - slip_cost

    trade_log = pd.DataFrame(index=df.index)
    trade_log['position'] = position
    trade_log['pnl'] = net_pnl
    trade_log['return_pct'] = net_pnl / capital
    trade_log['cum_return_pct'] = trade_log['return_pct'].cumsum()

    daily_ret = trade_log['return_pct'].dropna()
    ann_return = daily_ret.mean() * 252
    ann_vol = daily_ret.std() * np.sqrt(252)
    sharpe = ann_return / ann_vol if ann_vol > 0 else np.nan
    max_dd = (trade_log['cum_return_pct'] - trade_log['cum_return_pct'].cummax()).min()
    num_trades = ((position != 0) & (position.shift().fillna(0) == 0)).sum()

    return {
        'sharpe': sharpe,
        'cumulative_return': trade_log['cum_return_pct'].iloc[-1],
        'max_drawdown': max_dd,
        'trades': num_trades,
        'log': trade_log,
        'params': (z_entry, z_exit, z_window)
    }

# === WALK-FORWARD EVALUATION ===
def walk_forward_eval(data, beta, capital, slippage,
                      z_entry_range, z_exit_range, z_window_range,
                      window_months=6, step_months=3):

    all_test_logs = []
    summary = []

    dates = pd.to_datetime(data.index)
    start_date = dates.min()
    end_date = dates.max()

    while True:
        train_start = start_date
        train_end = train_start + pd.DateOffset(months=window_months)
        test_start = train_end + pd.Timedelta(days=1)
        test_end = test_start + pd.DateOffset(months=step_months)

        if test_end > end_date:
            break

        train_df = data.loc[train_start:train_end]
        test_df = data.loc[test_start - pd.Timedelta(days=30):test_end]

        best_result = None
        best_sharpe = -np.inf

        for z_entry, z_exit, z_window in product(z_entry_range, z_exit_range, z_window_range):
            if z_exit >= z_entry:
                continue
            res = backtest(train_df, z_entry, z_exit, z_window, beta, capital, slippage)
            if res['sharpe'] > best_sharpe:
                best_sharpe = res['sharpe']
                best_result = res

        test_res = backtest(test_df, *best_result['params'], beta, capital, slippage)
        test_log = test_res['log'].loc[test_start:test_end]
        test_log['walk'] = f"{test_start.date()} to {test_end.date()}"
        all_test_logs.append(test_log)

        summary.append({
            'train_start': train_start.date(),
            'train_end': train_end.date(),
            'test_start': test_start.date(),
            'test_end': test_end.date(),
            'z_entry': best_result['params'][0],
            'z_exit': best_result['params'][1],
            'z_window': best_result['params'][2],
            'sharpe': test_res['sharpe'],
            'cumulative_return': test_res['cumulative_return'],
            'max_drawdown': test_res['max_drawdown'],
            'trades': test_res['trades']
        })

        start_date += pd.DateOffset(months=step_months)

    all_logs_df = pd.concat(all_test_logs)
    return pd.DataFrame(summary), all_logs_df

# === PARAMETER GRID ===
z_entry_range = np.arange(1.0, 2.1, 0.2)
z_exit_range = np.arange(0.0, 1.0, 0.2)
z_window_range = np.arange(10, 31, 5)

# === RUN WALK-FORWARD ===
wf_summary, wf_logs = walk_forward_eval(
    data, beta_fixed, capital, slippage_rate,
    z_entry_range, z_exit_range, z_window_range
)

# === PLOT STITCHED TEST PERFORMANCE ===
plt.figure(figsize=(14, 6))
for label, group in wf_logs.groupby("walk"):
    plt.plot(group.index, group['cum_return_pct'] * 100, label=label)

plt.title("Walk-Forward Cumulative Return (%) by Test Period")
plt.xlabel("Date")
plt.ylabel("Cumulative Return (%)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

wf_summary
```

```python
# Script: pair_cointegration_backtest.py
# Purpose: Test multiple combinations of drone-related stocks for cointegration and backtest pairs trading

import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
from itertools import combinations
from datetime import datetime, timedelta

# === CONFIGURATION ===
tickers = ['CVX','SHEL','BP','USO','HES']  # Extendable list of drone-related stocks
lookback_period = "1y"
interval = "1d"
capital = 10000
slippage_rate = 0.001
z_entry = 1.5
z_exit = 0.2
z_window = 20
cointegration_pval_threshold = 0.15


# === FUNCTIONS ===

def get_data(ticker_list, period="1y", interval="1d"):
    data = yf.download(ticker_list, period=period, interval=interval)['Close']
    return data.dropna()

def test_cointegration(y, x):
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    hedge_ratio = model.params.iloc[1]
    resid = model.resid
    adf_test = sm.tsa.adfuller(resid)
    return {
        "hedge_ratio": hedge_ratio,
        "adf_stat": adf_test[0],
        "pval": adf_test[1],
        "crit_vals": adf_test[4],
        "resid": resid
    }

def backtest_pair(y, x, hedge_ratio, z_entry, z_exit, z_window, capital, slippage):
    spread = y - hedge_ratio * x
    mean = spread.rolling(z_window).mean()
    std = spread.rolling(z_window).std()
    zscore = (spread - mean) / std

    position = pd.Series(index=y.index, dtype=int)
    position[zscore > z_entry] = -1
    position[zscore < -z_entry] = 1
    position[zscore.abs() < z_exit] = np.nan
    position = position.ffill().fillna(0).astype(int)

    trade_value = y.abs() + hedge_ratio * x.abs()
    units = capital / trade_value
    spread_return = spread.diff()
    raw_pnl = position.shift(1) * spread_return * units.shift(1)
    signal = position.diff().fillna(0)
    slip_cost = signal.abs() * trade_value * slippage
    net_pnl = raw_pnl - slip_cost

    daily_ret = (net_pnl / capital).dropna()
    ann_return = daily_ret.mean() * 252
    ann_vol = daily_ret.std() * np.sqrt(252)
    sharpe = ann_return / ann_vol if ann_vol > 0 else np.nan
    max_dd = ((net_pnl / capital).cumsum() - (net_pnl / capital).cumsum().cummax()).min()

    return {
        "sharpe": sharpe,
        "cumulative_return": (net_pnl / capital).sum(),
        "max_drawdown": max_dd
    }

# === MAIN WORKFLOW ===

data = get_data(tickers, period=lookback_period, interval=interval)
valid_pairs = []

results = []

for t1, t2 in combinations(tickers, 2):
    y = data[t1]
    x = data[t2]

    try:
        cointegration = test_cointegration(y, x)
        print(f"{t1} and {t2} have pval {cointegration['pval']}")
        if cointegration['pval'] < cointegration_pval_threshold:
            perf = backtest_pair(y, x, cointegration['hedge_ratio'],
                                 z_entry, z_exit, z_window,
                                 capital, slippage_rate)
            results.append({
                'Ticker1': t1,
                'Ticker2': t2,
                'P-Value': cointegration['pval'],
                'ADF Stat': cointegration['adf_stat'],
                'Hedge Ratio': cointegration['hedge_ratio'],
                'Sharpe': perf['sharpe'],
                'Cumulative Return': perf['cumulative_return'],
                'Max Drawdown': perf['max_drawdown']
            })
    except Exception as e:
        print(f"Error processing pair ({t1}, {t2}): {e}")

results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by='Sharpe', ascending=False).reset_index(drop=True)
results_df
```
